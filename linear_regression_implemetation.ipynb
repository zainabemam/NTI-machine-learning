```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,2]
X
```

```python
# Standardize features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
import numpy as np

def linear_regression(X, y, learning_rate=0.01, n_iterations=1000):
    n_samples, n_features = X.shape
    # Initialize weights and bias
    weights = np.zeros(n_features)
    bias = 0

    for _ in range(n_iterations):
        # Linear prediction
        y_predicted = np.dot(X, weights) + bias

        # Compute gradients
        dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
        db = (1 / n_samples) * np.sum(y_predicted - y)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

    return weights, bias

def predict(X, weights, bias):
    return np.dot(X, weights) + bias
```

```python
from sklearn.linear_model import LinearRegression as SklearnLinearRegression



# Fit our custom linear regression using gradient descent
weights, bias = linear_regression(X, y)
custom_predictions = predict(X, weights, bias)

# Fit Scikit-Learn linear regression
sklearn_model = SklearnLinearRegression()
sklearn_model.fit(X, y)
sklearn_predictions = sklearn_model.predict(X)
```

```python
print(sklearn_predictions[0:5])
print(custom_predictions[0:5])
```

```python
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Calculate MSE
custom_mse = mean_squared_error(y, custom_predictions)
sklearn_mse = mean_squared_error(y, sklearn_predictions)

print("Custom Model MSE:", custom_mse)
print("Sklearn Model MSE:", sklearn_mse)
print()
```
